# Pandas profiling is a library that automatically generates the EDA for a dataset
## pandas_profiling has now changed to ydata_profiling

```python
import seaborn as sns
import pandas as pd
from ydata_profiling import ProfileReport
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
profile = ProfileReport(df, explorative=True)
profile.to_file('Tips.html')
```
