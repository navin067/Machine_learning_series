# K-Means clustering algorithm

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=1000, centers=3, n_features=2, random_state=23)
```

```python
X.shape
```

```python
plt.scatter(X[:,0], X[:,1])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.cluster import KMeans
```

```python
## Manual process to select the k-value

#Elbow method:

wcss = []
for k in range(1,11):
    Kmeans = KMeans(n_clusters=k, init='k-means++')
    Kmeans.fit(X_train)
    wcss.append(Kmeans.inertia_)
```

```python
wcss
```

```python
# plotting the elbow chart

plt.plot(range(1,11), wcss)
plt.xticks(range(1,11))
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

## From the chart, we select 3 as the required number of clusters as beyound it, the wcss seems to be stabalized.

```python
kmeans = KMeans(n_clusters=3, init='k-means++')
y_labels = kmeans.fit_predict(X_train)
```

```python
plt.scatter(X_train[:,0],X_train[:,1], c=y_labels)
```

```python
pred_labels = kmeans.predict(X_test)
plt.scatter(X_test[:,0],X_test[:,1], c=pred_labels)
```

## Performance metrics: Silhouette score

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_cofficients=[]
for k in range(2,11):
    kmeans = KMeans(n_clusters=k, init='k-means++')
    kmeans.fit(X_train)
    score = silhouette_score(X_train,kmeans.labels_)
    silhouette_cofficients.append(score)
```

```python
silhouette_cofficients
```

```python
#ploting the silhoutte_coeff with k values

plt.plot(range(2,11), silhouette_cofficients)
plt.xlabel('number of clusters')
plt.ylabel('silhouette accuracy')
plt.show()
```

# This also shows that for k=3, the silhoette score is the best.
