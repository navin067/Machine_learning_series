```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Pre-prunning is generally applied for larger dataset.

```python
df = sns.load_dataset('iris')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['species'] = le.fit_transform(df['species'])
```

```python
df.head()
```

```python
y = df.pop('species')
X = df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
```

## Applying pre-prunnning

```python
# setting hyperparameters value
from sklearn.model_selection import GridSearchCV
parameters = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'splitter': ['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features': ['auto', 'sqrt', 'log2']
}
dtcCV = GridSearchCV(dtc, param_grid=parameters, cv=5, scoring='accuracy')
```

```python
dtcCV.fit(X_train, y_train)
```

```python
dtcCV.best_params_
```

```python
dtcCV.best_score_
```

```python
preds = dtcCV.predict(X_test)
from sklearn.metrics import accuracy_score, classification_report
print(accuracy_score(preds, y_test))
print(classification_report(preds, y_test))
```

# The obtained accuracy 98% is really good. Here, 4 hyperparameters were selected using gridsearchcv during the model training phase which is known as pre-prunnin.
