## Practising some visualizations using seaborn

```python
import seaborn as sns

iris = sns.load_dataset('iris')
iris.head()
```

```python
iris.dtypes
```

Contains 4 numeric values and one categorical value

```python
iris.describe()
```

```python
iris.shape
```

```python
# finding unique values in each column of the dataset
for col in iris.columns:
    print(iris[col].nunique())
```

## distribution plots for numeric features

distplot
joinplot
pairplot

```python
import matplotlib.pyplot as plt


plt.subplot(2,2,1)
sns.histplot(data=iris, x='sepal_length')
plt.subplot(2,2,2)
sns.histplot(data=iris, x='sepal_width')
plt.subplot(2,2,3)
sns.histplot(data=iris, x='petal_length')
plt.subplot(2,2,4)
sns.histplot(data=iris, x='petal_width')

plt.show()
```

```python
sns.jointplot(x='sepal_length', y='sepal_width', data=iris, kind='hex')
```

```python
sns.jointplot(x='sepal_length', y='sepal_width', data=iris, kind='scatter', hue='species')
```

```python
sns.jointplot(x='petal_length', y='petal_width', data=iris, kind='scatter', hue='species')
```

```python
sns.pairplot(data=iris, hue='species')
```

```python
df = iris.drop('species', axis=1)
df.corr()
```

```python
sns.heatmap(df.corr(), annot=True)
```

## categorical plots

```python
sns.countplot(data=iris,x='species')
```

```python
sns.boxplot(data=iris, x='species', y='sepal_length')
```

```python
sns.violinplot(data=iris, y='species', x='sepal_length')
```
