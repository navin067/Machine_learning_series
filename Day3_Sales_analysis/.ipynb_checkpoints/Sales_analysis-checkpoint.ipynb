```python
import math
months = df['Month'].unique()
products = df['Product'].unique()

month_record = df.groupby(['Product','Month'])[['Quantity Ordered','Price Each']].sum()

plt.figure(figsize=(20,25))
x=1
for product in products:
    df_trend = month_record.loc[(product, months),['Quantity Ordered','Price Each']]
    plt.subplot(math.ceil(len(products)/3),3,x)  
    plt.plot(months, df_trend['Quantity Ordered'].values, marker='o',label='total_quantity') #line plot for clicks
    plt.plot(months, df_trend['Price Each'].values, marker='o',label='revenue')
    plt.grid(alpha=0.45)
    plt.title(f"Revenue and order trends for {product}")
    plt.yscale('log')
    x+=1
```

# Answering some sale analysis questions.

### Question 1: What was the best month for sales ? How much was earned that month ?

```python
df.head()
```

```python
# for answering the question we have to group by the dataframe based on months
# and then find the sum of price for all the products sold
# also we have to add the total sales which is obtiained by quantity and price each

df['Total_sales'] = df['Quantity Ordered'] * df['Price Each']
df.head()
```

```python
revenue_per_month = pd.DataFrame(df.groupby('Month')['Total_sales'].sum())
type(revenue_per_month)
```

```python
revenue_per_month.sort_values('Total_sales', ascending=False)
```

```python
revenue_per_month['Month'] = revenue_per_month.index
sns.barplot(data=revenue_per_month, x='Month', y='Total_sales')
plt.title('Total sales across different months')
plt.show()
```

### This shows that the month december has highest sales.

## What city had the highest number of sales ?

```python
df.head()
```

```python
str1 = '944 Walnut St, Boston, MA 02215'
str1.split(r', ')[1]
```

```python
# adding a city column into the dataframe for answering the quesiton

df[['Home','City','Postal code']] = df['Purchase Address'].str.split(r', ', expand=True)
df.head()
```

```python
city_records = pd.DataFrame(df.groupby('City')['Total_sales'].sum())
city_records.sort_values('Total_sales', ascending=False)
```

```python
city_records.sort_values('Total_sales', ascending=False).plot.barh()
plt.title('Total sales per city in 2019')
plt.grid(alpha=0.25)
plt.show()
```

```python
df.drop(['Home','City','Postal code'], axis=1, inplace=True)
```

```python
df.head()
```

```python
# other way to get city is 

df['City'] = df['Purchase Address'].apply(lambda x: x.split(', ')[1])
df.head()
```

```python
# to get state name

str1 = '944 Walnut St, Boston, MA 02215'
str1.split(',')[2].split(' ')[1]
```

```python
def get_city(address):
    return address.split(',')[1]
def get_state(address):
    return address.split(',')[2].split(' ')[1]

df['City'] = df['Purchase Address'].apply(lambda x: f" {get_city(x)}  ({get_state(x)})")
df.head()
```

```python
city_records = pd.DataFrame(df.groupby('City')['Total_sales'].sum())
city_records.sort_values('Total_sales', ascending=False).plot.barh()
plt.title('Total sales per city in 2019')
plt.grid(alpha=0.25)
plt.xlabel('Log scaled sales amount')
plt.show()
```

## What time should we display advertisements to maximize likelihood of customer's buying product ?

```python
df.dtypes
```

```python
df['Order Date']
```

```python
df['Hour'] = df['Order Date'].dt.hour
df['Minute'] = df['Order Date'].dt.minute
df.head()
```

```python
df['Hour'].value_counts()
```

```python
hourly_sales = pd.DataFrame(df.groupby('Hour')['Total_sales'].sum())
hourly_sales.nlargest(6, 'Total_sales')
```

```python
plt.figure(figsize=(12,6))

hours = [hour for hour,df in df.groupby('Hour')]
plt.plot(hours, df.groupby('Hour')['Order ID'].count().values, marker='o')
plt.grid(alpha=0.35)
plt.xticks(hours)
plt.title('Order count for each hour')
plt.xlabel('Hours')
plt.ylabel('Order count')
plt.show()
```

### Peak time people making orders is around 11am to 1pm and at 7pm.

```python
new_df = df[df['Order ID'].duplicated(keep=False)]
new_df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

new_df = new_df[['Order ID', 'Grouped']].drop_duplicates()
```

```python
new_df.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in new_df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

#count.most_common(10)

for key, value in count.most_common(10):
    print(key, value)
```

## What product sold the most? Why do you think it sold the most?

```python
df.head()
```

```python
sold_products = pd.DataFrame(df.groupby('Product')['Quantity Ordered'].sum()).reset_index()
sold_products.sort_values('Quantity Ordered',ascending=False)
```

```python
sold_products.nlargest(10, 'Quantity Ordered').plot.bar()
plt.title('Top 10 most sold products')
plt.ylabel('Quantities sold')
plt.show()
```

```python
price_per_product = df[['Product','Price Each']].set_index('Product').to_dict()
price_per_product['Price Each']

sold_products['Price Each'] = sold_products['Product'].map(price_per_product['Price Each'])
```

```python
plt.figure(figsize=(12,6))
plt.bar(sold_products['Product'], sold_products['Quantity Ordered'])
plt.xticks(rotation=90)
plt.title('Products and their orders')
plt.xlabel('Name of the products')
plt.ylabel('Quantity Ordered')
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(12,6))
ax2 = ax1.twinx()
ax1.bar(sold_products['Product'], sold_products['Quantity Ordered'])
ax2.plot(sold_products['Product'].values,sold_products['Price Each'].values, 'r-')
ax1.set_xlabel('Products')
ax1.set_ylabel('Quantity Ordered')
ax2.set_ylabel('Price')
ax1.set_xticklabels(sold_products['Product'], rotation=90, size=8)

plt.show()
```
